## Step 0: Import libraries

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

```

## Step 1: Import data

### ACEI and clusters similarity 


```python

data_ace = pd.read_csv('/Users/odilehasa/Hypertension/Final_Experiments/FINAL - October/Output/6.3 Similarity_Cluster 2 and ACEI.csv')

data_ace
```

### ARB and clusters similarity 


```python

data_arb = pd.read_csv('/Users/odilehasa/Hypertension/Final_Experiments/FINAL - October/Output/6.4. Similarity_Cluster 0 and ARB.csv')

data_arb
```

## Step 2: Create dataframe with only ACEI and ARB-like clusters (as per phase 6.2)

```python
# ACEIs
# For example, in phase 6.2, cluster 0 was identified as ARBs and cluster 2 as ACEIs

df_ACEIsim = data_ace[data_ace['Similarity Score_2ACEI'] >= 0.7]
df_ACEIsim
```

```python
# ARBs
# For example, in phase 6.2, cluster 0 was identified as ARBs and cluster 2 as ACEIs

df_ARBsim = data_arb[data_arb['Similarity Score_0ARB'] >= 0.7]
df_ARBsim
```

```python
# Create dataframe with ACEI and ARB similarity scores

sim_data = [df_ACEIsim['Similarity Score_2ACEI'], df_ARBsim['Similarity Score_0ARB']]

column_headers = ["Similarity Score_2ACEI", "Similarity Score_0ARB"]

df_sim = pd.concat(sim_data, axis = 1, keys = column_headers)

df_sim
```

```python
# set figure size 
fig, ax_hist = plt.subplots(figsize=(20, 10))
plt.rcParams.update({'font.size': 20})   # specify font size for all labels without specified font sizes

# draw histogram
ax_hist = sns.histplot(data= df_sim, multiple= "stack", palette= "mako", kde = True)

# specify axis labels
ax_hist.set_xlabel('Tanimoto Coefficient (Tc)', fontsize=25)
ax_hist.set_ylabel('Count', fontsize=25)

# save figure
plt.savefig('SUPP 0. Tanimoto similarity scores of lead molecule.png')
```
