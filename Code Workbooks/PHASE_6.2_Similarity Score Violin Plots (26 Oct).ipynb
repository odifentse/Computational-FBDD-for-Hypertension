The purpose of this notebook is to take the similarity scores of the new molecules and target molecules to visualise this as violin plots.

This will enable us to see which clusters are most similar to the difference drug classes. The cluster with the highest score in each drug class will be considered to contain molecules resembling those of that respective drug class.

## Step 0: Import the necessary libraries

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

```

## Step 1: Import Data

```python
# ACEI and clusters similarity 

data_ace = pd.read_csv('/Users/odilehasa/Hypertension/Final_Experiments/FINAL - October/Output/6. ACEIs all cluster similarity scores.csv', index_col=0)
data_ace
```

```python
# ARB and clusters similarity 

data_arb = pd.read_csv('/Users/odilehasa/Hypertension/Final_Experiments/FINAL - October/Output/6. ARBs all cluster similarity scores.csv',index_col=0)
data_arb
```

## Step 2: Draw Violin Plots original molecules vs Clusters

### ACEI vs Clusters

```python
# set figure size 
fig, ax_ace = plt.subplots(figsize =(20, 10))
plt.rcParams.update({'font.size': 20})   # specify font size for all labels without specified font sizes

# draw violin plot
ax_ace = sns.violinplot(data = data_ace, palette = "Accent")

# add axis titles
ax_ace.set_title('Similarity scores of new molecule clusters against original ACEIs', fontsize = 25)
ax_ace.set_ylabel('Similarity Score', fontsize = 25)

# save figure
plt.savefig('6.7. Similarity scores of new molecule clusters against original ACEIs.png')
```

### ARBs vs Clusters

```python

# set figure size 
fig, ax_arb = plt.subplots(figsize= (20, 10))
plt.rcParams.update({'font.size': 20})   # specify font size for all labels without specified font sizes

# draw violin plot
ax_arb =sns.violinplot(data = data_arb, palette = "YlOrBr_r", gridsize = 800, width = 0.8)

# add axis titles
ax_arb.set_title('Similarity scores of new molecule clusters against original ARBs', fontsize=25)
ax_arb.set_ylabel('Similarity Score', fontsize=25)

# save figure
plt.savefig('6.8. Similarity scores of new molecule clusters against original ARBs.png')
```
